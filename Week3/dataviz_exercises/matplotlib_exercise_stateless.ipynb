## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
```

```python
df.head()

# df.columns
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
x = df['month_number']
y = df['total_profit']


fig, ax = plt.subplots(figsize = (10,6))

ax.set_xlabel('Month')
ax.set_ylabel('Total Profit')
ax.set_title('Total Profit per Month')
ax.plot(x,y,'red')
plt.show()
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']
months = df['month_number']

x = df['month_number']

# create fig and axes, set the fig size
fig, ax = plt.subplots(figsize = (10,6))

# plots
ax.plot(x, facecream, label = 'facecream')
ax.plot(x, facewash, label = 'facewash')
ax.plot(x, toothpaste, label = 'toothpaste')
ax.plot(x, bathingsoap, label = 'bathingsoap')
ax.plot(x, shampoo, label = 'shampoo')
ax.plot(x, moisturizer, label = 'moisturizer')

# set legend and axes
ax.legend(loc= 'upper left')
ax.set_xlabel('Month')
ax.set_ylabel('Units sold')

# get handles and labels
handles, labels = plt.gca().get_legend_handles_labels()

# specify order of items in legend
order = [3, 2, 0, 4, 1]

ax.legend([handles[idx] for idx in order], [labels[idx] for idx in order])
plt.show()
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
# get total profit
x = df['total_profit']

# create fig
fig, ax = plt.subplots(figsize = (10,6))

ax.hist(x,bins=12)
plt.show()
```

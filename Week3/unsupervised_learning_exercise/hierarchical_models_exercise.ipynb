## Hierachical Models Assignment

```python
# import pandas
import pandas as pd

# import Agglomerative clustering
from sklearn.cluster import AgglomerativeClustering

# import scipy.cluster.hierarchy
import scipy.cluster.hierarchy as sch

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt

# set figure size
plt.rcParams["figure.figsize"] = (12,8)
```

```python
# Load dataframe
df = pd.read_csv('https://raw.githubusercontent.com/lucko515/clustering-python/master/Customer%20in%20Mall%20clusterng/Mall_Customers.csv')
```

```python
df.head()
```

## Objective: 

- Fit Hierarchical clustering, 
- find optimal number of clusters via dendrogram 
- and plot clusters. 

Again, to simplify plotting in 2-D, we can create clustering model based on the last two columns of the dataset only.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_array = df[['Annual Income (k$)', 'Spending Score (1-100)']].to_numpy()

scaled_data = scaler.fit_transform(data_array)
```

* define function plot_dendogram

```python
def plot_dendrogram(X, method = 'ward'):
    dendrogram = sch.dendrogram(sch.linkage(X, method = method))
    plt.title('Dendrogram')
    plt.ylabel('Euclidean Distance')
    plt.xlabel('Points')
    plt.show()
```

* Plot dendrogram

```python
plot_dendrogram(scaled_data)


# from dendrogram, im guessing there are 3 clusters because the longest vertical line is
# the left one and drawing horizontal lines across at extremities
```

* Fit Hierarchical clustering with optimal number of clusters

```python
ac = AgglomerativeClustering(affinity = 'euclidean',
                            linkage = 'ward',
                            n_clusters = 3)

y_hc = ac.fit_predict(scaled_data)
```

* define function plot_clusters

```python
def plot_clusters(X, y_res, plt_cluster_centers = False):
    X_centroids = []
    Y_centroids = []

    for cluster in set(y_res):
        x = X[y_res == cluster, 0]
        y = X[y_res == cluster, 1]
        X_centroids.append(np.mean(x))
        Y_centroids.append(np.mean(y))

        plt.scatter(x, y, s = 50, marker= 's', label = f'cluster {cluster}')
        
    if plt_cluster_centers:
        plt.scatter(X_centroids,
                    Y_centroids,
                    marker = '*',
                    c = 'red',
                    s = 250,
                    label = 'centroids')
    
    plt.legend()
    plt.grid()
    plt.show()
```

* Plot HC clusters

```python
plot_clusters(scaled_data, y_hc)
```
